```python
import sys
sys.path.append('drive/My Drive/CCTech/Model/BoundariesBranch/Scripts/')
import numpy as np
import tensorflow
import tensorflow.keras
import paperLayersModel
from paperLayersModel import *
import setLearningRate
from setLearningRate import *
import dataGen
from dataGen import *
import random
```

```python
train_input = np.load('drive/My Drive/CCTech/MasksGeneration/Masks/train_inputs.npy')
train_target = np.load('drive/My Drive/CCTech/MasksGeneration/Masks/train_bmasks.npy')
test_input = np.load('drive/My Drive/CCTech/MasksGeneration/Masks/test_inputs.npy')
test_target = np.load('drive/My Drive/CCTech/MasksGeneration/Masks/test_bmasks.npy')
```

```python
train_generator = DataGenerator(train_input, train_target, batch_size=1, dim=(512,512), n_channels=3)
validation_generator = DataGenerator(test_input, test_target, batch_size=1, dim=(512,512), n_channels=3)
```

```python
model = paperLayersModel()
```

```python
model.load_weights("drive/My Drive/CCTech/Model/BoundariesBranch/BoundaryWeights.h5")
```

```python
train_pred = []
for i in range(len(train_input)):
  print(i)
  #find prediction of ith train sample
  y = model.predict(train_input[i])
  train_pred.append(y)
  #find true and predicted train images
  im_train_true = train_target[i][0]
  im_train_pred = y[0]
  path_true = "drive/My Drive/CCTech/Model/BoundariesBranch/Predictions/Train/True/" + str(i) + ".jpg"
  path_pred = "drive/My Drive/CCTech/Model/BoundariesBranch/Predictions/Train/Predicted/" + str(i) + ".jpg"
  tensorflow.keras.preprocessing.image.save_img(path_true,im_train_true)
  tensorflow.keras.preprocessing.image.save_img(path_pred,im_train_pred)

np.save("drive/My Drive/CCTech/Model/BoundariesBranch/Predictions/Train/train_pred.npy",train_pred)
```

```python
test_pred = []
for i in range(len(test_input)):
  print(i)
  #find prediction of ith train sample
  y = model.predict(test_input[i])
  test_pred.append(y)
  #find true and predicted train images
  im_test_true = test_target[i][0]
  im_test_pred = y[0]
  path_true = "drive/My Drive/CCTech/Model/BoundariesBranch/Predictions/Test/True/" + str(i) + ".jpg"
  path_pred = "drive/My Drive/CCTech/Model/BoundariesBranch/Predictions/Test/Predicted/" + str(i) + ".jpg"
  tensorflow.keras.preprocessing.image.save_img(path_true,im_test_true)
  tensorflow.keras.preprocessing.image.save_img(path_pred,im_test_pred)

np.save("drive/My Drive/CCTech/Model/BoundariesBranch/Predictions/Test/test_pred.npy",test_pred)
```
